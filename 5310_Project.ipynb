```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Define the connection URL. General form
conn_url = 'postgresql://postgres:Yeqianchi123@localhost/5310_Project'

#Create an engine in order to connects to PostgreSQL
engine = create_engine(conn_url)

#Establish the connection
connection = engine.connect()
```

```python
#Define the CSV file paths and their corresponding table names
csv_files_and_tables = {
    '/Users/yeqianchi/Desktop/5310/csv/offices.csv': 'offices',
    '/Users/yeqianchi/Desktop/5310/csv/departments.csv': 'departments',
    '/Users/yeqianchi/Desktop/5310/csv/houses.csv': 'houses',
    '/Users/yeqianchi/Desktop/5310/csv/clients.csv': 'clients',
    '/Users/yeqianchi/Desktop/5310/csv/employees.csv': 'employees',
    '/Users/yeqianchi/Desktop/5310/csv/transactions.csv': 'transactions',
    '/Users/yeqianchi/Desktop/5310/csv/client_feedbacks.csv': 'client_feedbacks',
    '/Users/yeqianchi/Desktop/5310/csv/salaries.csv': 'salaries',
    '/Users/yeqianchi/Desktop/5310/csv/expenses.csv': 'expenses',
    '/Users/yeqianchi/Desktop/5310/csv/profits.csv': 'profits',
    '/Users/yeqianchi/Desktop/5310/csv/sales.csv': 'sales',
    '/Users/yeqianchi/Desktop/5310/csv/house_amenities.csv': 'house_amenities',
    '/Users/yeqianchi/Desktop/5310/csv/house_records.csv': 'house_records',
    '/Users/yeqianchi/Desktop/5310/csv/market_trends.csv': 'market_trends',
    '/Users/yeqianchi/Desktop/5310/csv/marketing_activities.csv': 'marketing_activities',
    '/Users/yeqianchi/Desktop/5310/csv/appointments.csv': 'appointments'
}
```

```python
# Dictionary to store DataFrames for displaying head()
dataframes = {}

# Extract and Load
for csv_file, table_name in csv_files_and_tables.items():
    try:
        # Extract: Read CSV files
        df = pd.read_csv(csv_file)
        
        # Store the DataFrame for loading
        dataframes[table_name] = df
        
        # Load: Write data to the corresponding table in the database
        df.to_sql(table_name, engine, if_exists='append', index=False)  # Use append to avoid dropping the table structure
        print(f'Data from {csv_file} has been successfully imported into {table_name} table in 5310_Project database.')
            
    except FileNotFoundError as e:
        print(f'File not found: {csv_file}')
    except Exception as e:
        print(f'Error processing {csv_file}: {e}')
```

```python
#display the head of each table in the database
for table_name, df in dataframes.items():
    print(f'Head of {table_name} table:')
    display(df.head())
    print('\n')
```
